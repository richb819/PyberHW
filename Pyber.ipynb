```python
import pandas as pd 
citydata= pd.read_csv("city_data.csv")
citydata.head()
```

```python
citydata.head(10)
```

```python
ridedata= pd.read_csv("ride_data.csv")
ridedata.head()
```

```python
driver_count=[citydata[citydata.city==row.city].iloc[0]["driver_count"] for index,row in ridedata[["city"]].iterrows()]
ridedata["driver_count"]=driver_count

city_type=[citydata[citydata.city==row.city].iloc[0]["type"] for index,row in ridedata[["city"]].iterrows()]
ridedata["type"]=city_type

ridedata.head()
```

```python
Sarabury_data=ridedata[ridedata.city=="Sarabury"]
numrides=len(Sarabury_data)
averagefare=Sarabury_data["fare"].mean()
print("Number of Rides:",numrides,"average fare:",averagefare)
```

```python
num_rides=[len(ridedata[ridedata.city==row.city]) for index, row in citydata.iterrows()]
citydata["num_rides"]=num_rides
average_fare=[ridedata[ridedata.city==row.city].fare.mean() for index, row in citydata.iterrows()]
citydata["average_fare"]=average_fare
citydata.head()
```
